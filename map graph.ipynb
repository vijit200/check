```python
df1 = pd.read_csv('https://raw.githubusercontent.com/nikhilkumarsingh/choropleth-python-tutorial/master/india_census.csv')
```

```python
df1
```

```python
d = requests.get('https://raw.githubusercontent.com/nikhilkumarsingh/choropleth-python-tutorial/master/states_india.geojson')
```

```python
data = d.json()
```

```python
data['features'][0]['properties']
```

```python
a = {}
for feature in data['features']:
    feature['id'] = feature['properties']['state_code']
    a[feature['properties']['st_nm']] = feature['id']
```

```python
a
```

```python
df1['id'] = df1['State or union territory'].apply(lambda x:a[x])
```

```python
df1
```

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'browser'
```

```python
fig = px.choropleth(df1,locations='id',geojson=data,color = 'Population',hover_data=['Population'],hover_name='State or union territory')
fig.update_geos(fitbounds = 'locations',visible = False)
fig.show()
```
