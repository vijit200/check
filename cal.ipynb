```python
from tkinter import *
from tkinter import messagebox
from tkinter import simpledialog
import tkinter.font as TkFont

class cal():
    def __init__(self,root):
        self.root = root
        self.root.geometry('250x300')
        self.my_font = TkFont.Font(size=10,weight='bold')
        self.my_font1 = TkFont.Font(weight='bold')
        self.l1 = Label(root,text = 'Calculator',font='Arial 18 bold',bg = 'yellow',fg = '#FF33EF')
        self.l2 = Label(root,text = 'First_No',bg = '#58D68D',font=self.my_font1,fg = '#4233FF')
        self.l3 = Label(root,text = 'Second_No',bg = '#58D68D',font =self.my_font1,fg = '#4233FF')
        self.l4 = Label(root,text = 'Result',bg = '#58D68D',font=self.my_font1,fg = '#4233FF')
        self.e1 = Entry(root,fg='red')
        self.e2 = Entry(root,fg = 'red')
        self.e3 = Entry(root,fg = 'blue')
        self.b1 = Button(root,text = 'ADD',height=3,width=10,command=self.add,bg = '#EDBB99',font=self.my_font,fg = 'red')
        self.b2 = Button(root, text = 'SUBTRACT',height=3,width=10,command=self.sub,bg = '#EDBB99',font=self.my_font,fg = 'red')
        self.b3 = Button(root,text = 'MULTIPLY',height=3,width=10,command=self.mul,bg = '#EDBB99',font=self.my_font,fg = 'red')
        self.b4 = Button(root,text = 'DIVIDE',height=3,width=10,command=self.div,bg = '#EDBB99',font=self.my_font,fg = 'red')
        self.c1 = Button(root,text = 'clear',command=self.clear,bg = '#5D6D7E',font=self.my_font1)
        self.q1 = Button(root,text = 'Quit',command=self.quit,bg = '#5D6D7E',font=self.my_font1)
        self.c1.grid(row = 6 , column=0)
        self.q1.grid(row= 6,column=1)
        self.l1.grid(row = 0 ,column=1)
        self.l2.grid(row = 1 ,column=0)
        self.e1.grid(row = 1 ,column=1)
        self.l3.grid(row = 2,column=0)
        self.e2.grid(row = 2 ,column=1)
        self.l4.grid(row =3,column=0 )
        self.e3.grid(row = 3,column=1)
        self.b1.grid(row = 4 , column= 0,pady=5, padx=5, sticky=E + W)
        self.b2.grid(row=4, column=1,pady=5, padx=5, sticky=E + W)
        self.b3.grid(row=5, column=0,pady=5, padx=5, sticky=E + W)
        self.b4.grid(row=5, column=1,pady=5, padx=5, sticky=E + W)
        root.config(bg = 'lightblue')
        root.resizable(False,False)
        self.root.mainloop()

    def add(self):
        try:
            self.e3.delete(0,END)
            self.x = int(self.e1.get())
            self.y = int(self.e2.get())
            self.s = self.x + self.y
            self.e3.insert(0,str(self.s))
        except Exception :
            messagebox.showerror('ERROR','only digit is allow')

    def sub(self):
        try:
            self.e3.delete(0,END)
            self.x = int(self.e1.get())
            self.y = int(self.e2.get())
            self.s = self.x - self.y
            self.e3.insert(0,str(self.s))
        except Exception :
            messagebox.showerror('ERROR','only digit is allow')

    def mul(self):
        try:
            self.e3.delete(0,END)
            self.x = int(self.e1.get())
            self.y = int(self.e2.get())
            self.s = self.x * self.y
            self.e3.insert(0,str(self.s))
        except Exception :
            messagebox.showerror('ERROR','only digit is allow')
    def div(self):
        try:
            self.e3.delete(0,END)
            self.x = int(self.e1.get())
            self.y = int(self.e2.get())
            self.s = self.x / self.y
            self.e3.insert(0,str(self.s))
        except Exception :
            messagebox.showerror('ERROR','only digit is allow')

    def clear(self):
            self.e3.delete(0, END)
            self.e2.delete(0, END)
            self.e1.delete(0, END)

    def quit(self):
        s = messagebox.askyesno('Quiting','Are you sure you want to quit!')
        if s == True:
            name = simpledialog.askstring('Input','What is your name')
        if name is None or name == '':
            name = 'UserJee!'
        messagebox.showinfo('Thank You','Have a good day {}'.format(name))
        self.root.destroy()

root = Tk()
s = cal(root)
s.__init__(root)
```
